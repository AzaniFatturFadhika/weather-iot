```python
import requests
import pandas as pd

url = "https://archive-api.open-meteo.com/v1/archive"
params = {
    "latitude": -7.0520702239386175, 
    "longitude": 110.43532807750137,
    "start_date": "2000-01-01",
    "end_date": "2025-11-27",
    "hourly": "temperature_2m,relative_humidity_2m,pressure_msl,precipitation,wind_speed_10m",
    "timezone": "auto"
}

response = requests.get(url, params=params)
data = response.json()

df = pd.DataFrame({
    "time": data["hourly"]["time"],
    "temperature": data["hourly"]["temperature_2m"],
    "humidity": data["hourly"]["relative_humidity_2m"],
    "pressure": data["hourly"]["pressure_msl"],
    "rain": data["hourly"]["precipitation"],
    "wind_speed": data["hourly"]["wind_speed_10m"]
})

df.to_csv("semarang_weather_2000_2025.csv", index=False)
print("Data Semarang berhasil disimpan!")
```
